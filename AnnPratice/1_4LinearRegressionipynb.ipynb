3️⃣ Calculate the Gradient

Write the loss function clearly:

L = (1/2) * (y - (w*x + b))^2

Now take the derivative with respect to w:

dL/dw

Derivative result:

dL/dw = -(y - y_pred) * x

But we know:

y - y_pred = error

So the derivative becomes:

dL/dw = - error * x

```python
import numpy as np

X = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])

w = 0
b = 0
lr = 0.01
epochs = 1000

for j in range(epochs):

    for i in range(len(X)):

        y_pred = w*X[i] + b

        error = y[i] - y_pred

        w = w + lr * error * X[i]

        b = b + lr * error

print("Weight:", w)
print("Bias:", b)

x_test = 10
print("Prediction:", w*x_test + b)
```
